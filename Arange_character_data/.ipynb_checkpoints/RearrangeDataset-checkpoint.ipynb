```python
import glob
base_folder='C:\\Users\\Ajay\\Documents\\Code\\EXL\\Projects\\PaperlessAuditPOC\\Arange_character_data'
data_folder='\\V0.3\\data'
extract_to_folder=r'\handwritten_alphabets'
final_res=[]
for i in main_dir_list:
    dir_path=base_folder+data_folder+"\\"+i+"\\*.*"
    
    res = glob.glob(dir_path)
    final_res.append(res)
```

```python
final_res
```

```python
extract_to_folder="\\handwritten_alphabets"
try:
    os.mkdir(base_folder+extract_to_folder)
except FileExistsError:
    print("folder already exists")
import shutil
i=1
for i in range(1,41):
    new_folder=extract_to_folder+"\\person"+str(i)
    try:
        os.mkdir(base_folder+new_folder)
    except FileExistsError:
        print(base_folder+new_folder," already exists")
for files in final_res:
    for i in range(0,40):
        
        new_folder=extract_to_folder+"\\person"+str(i)
        try:
            
            shutil.copy(files[i], base_folder+new_folder)
        except FileExistsError:
            shutil.copy(files[i], base_folder+new_folder)
        i+=1
```
